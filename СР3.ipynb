```python
#  импортируем 
from ludwig.api import LudwigModel
```

```python
# задаем параметры модели 
model_definition = {
    "input_features": [{'name': 'title', 'type': 'text', 'level': 'word'},
                       {'name': 'location', 'type': 'text', 'level': 'word'},
                       {'name': 'department', 'type': 'category'},
                       {'name': 'salary_range', 'type': 'category'},
                       {'name': 'company_profile', 'type': 'text', 'level': 'word'},
                       {'name': 'description', 'type': 'text', 'level': 'word'},
                       {'name': 'requirements', 'type': 'text', 'level': 'word'},
                       {'name': 'benefits', 'type': 'text', 'level': 'word'},
                       {'name': 'telecommuting', 'type': 'category'},
                       {'name': 'has_company_logo', 'type': 'category'},
                       {'name': 'has_questions', 'type': 'category'},
                       {'name': 'employment_type', 'type': 'category'},
                       {'name': 'required_experience', 'type': 'category'},
                       {'name': 'required_education', 'type': 'category'},
                       {'name': 'industry', 'type': 'category'},
                       {'name': 'function', 'type': 'category'},
    ],
    "output_features": [
        {'name': 'fraudulent', 'type': 'category'}
    ]
}
```

```python
# обучаем модель
model = LudwigModel (model_definition)
```

```python
train_stats = model.train(df)
```
